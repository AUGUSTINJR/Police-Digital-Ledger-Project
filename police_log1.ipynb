```python
#Data cleaning
#Due to Categorical Data we choose mode method
import numpy as np
df['search_type'].replace([' ','','nan','NaN'],np.nan)
df['search_type'].fillna(df['search_type'].mode()[0],inplace=True)
```

```python
#change stop_time datatype
df['stop_time'] = pd.to_datetime(df['stop_time'], format='%H:%M:%S').dt.time
```

```python
#create timestamp 
df['timestamp']=pd.to_datetime(df['stop_date'].astype(str) + ' ' +df['stop_time'].astype(str))
```

```python
insert_query="""
    INSERT INTO police_Postlog(
    stop_date,stop_time,country_name,driver_gender,driver_age_raw,driver_age,driver_race,violation_raw,
    violation,search_conducted,search_type,stop_outcome,is_arrested,stop_duration,drugs_related_stop,vehicle_number,timestamp
    ) VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
"""
```

```python
for _, row in df.iterrows():
    mycursor.execute(insert_query,(
        row['stop_date'],
        row['stop_time'],
        row.get('country_name'),
        row.get('driver_gender'),
        int(row.get('driver_age_raw')),
        int(row.get('driver_age')),
        row.get('driver_race'),
        row.get('violation_raw'),
        row.get('violation'),
        bool(row.get('search_conducted')),
        row.get('search_type'),
        row.get('stop_outcome'),
        bool(row.get('is_arrested')),
        row.get('stop_duration'),
        bool(row.get('drugs_related_stop')),
        row.get('vehicle_number',f'VEH(row.name)'),
        row['timestamp']
    ))
```
